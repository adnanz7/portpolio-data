# Data Order Penjualan

```python
import pandas as pd

data_penjualan = pd.read_csv('order dqlab.csv')
data_penjualan
```

## Tampilkan Kolom

```python
data_penjualan.columns
```

## Tampilkan Jenis Pembayaran

```python
data_penjualan.payment_type.value_counts().to_frame()
```

## Tampilkan Persentase Jenis Pembayaran

```python
pjp = data_penjualan.payment_type.value_counts(normalize = True)*100
pjp.to_frame()
```

## Tampilkan Jenis Pembayaran Berdasarkan Kategori Produk

```python
data_penjualan.groupby('product_category_name').payment_type.value_counts().to_frame()
```

## Tampilkan Persentase Jenis Pembayaran Berdasarkan Kategori Produk

```python
data_penjualan.groupby('product_category_name').payment_type.value_counts(normalize = True).to_frame()*100
```

## Tampilkan Jumlah Penjualan Berdasarkan Kategori Produk

```python
data_penjualan.groupby('product_category_name').price.sum().to_frame()
```

## Tampilkan Persentase Jumlah Penjualan Berdasarkan Kategori Produk

```python
sp = data_penjualan.groupby('product_category_name').price.sum()/data_penjualan.price.agg(sum) * 100
sp.to_frame()
```
