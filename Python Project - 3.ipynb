# Data Penjualan

```python
import pandas as pd
penjualan = pd.read_excel('sample_pivot.xlsx')
penjualan
```

## Tampilkan Penjumlahan Unit & Sales berdasarkan Type Clothing

```python
pd.pivot_table(penjualan, index = 'Type', aggfunc = 'sum')
```

## Tampilkan Penjumlahan Unit & Sales berdasarkan Region

```python
pd.pivot_table(penjualan, index = 'Region', aggfunc = 'sum')
```

## Tampilkan Penjumlahan Sales & Units Berdasarkan Region & Type Clothing

```python
pd.pivot_table(penjualan, index = 'Type', columns = 'Region', aggfunc = 'sum', values = ['Sales', 'Units'], margins = True, margins_name = 'Total')
```
